```python
import tensorflow as tf
import keras
```

```python
inputs0 = keras.Input(shape=(4, ))
h0_0 = keras.layers.Dense(8, activation="relu")(inputs0)

inputs1 = keras.Input(shape=(6, ))
h0_1 = keras.layers.Dense(4, activation="relu")(inputs1)

inputs2 = keras.Input(shape=(2, ))

h1 = keras.layers.Concatenate()([h0_0, h0_1, inputs2])

outputs0 = keras.layers.Dense(8)(h1)
outputs1 = keras.layers.Dense(2)(h1)

model = keras.Model(
    inputs=[inputs0, inputs1, inputs2],
    outputs=[outputs0, outputs1],
)
```

```python
keras.utils.plot_model(model, show_shapes=True)
```

```python
x0 = tf.random.normal([3, 4])
x1 = tf.random.normal([3, 6])
x2 = tf.random.normal([3, 2])

y = model([x0, x1, x2])
```

Functional API를 이용해 output이 여러 개인 모델을 만들어 봅시다.   
   
만들어야 할 모델 구조는 아래 그림과 같습니다. 각 output 직전의 Dense 레이어는    
활성화 함수를 사용하지 않고, 나머지 Dense 레이어는 활성화 함수로 ReLU를 사용해 주세요.

```python
#import tensorflow as tf
#import keras

inputs = keras.Input(shape=(4, ))

# 여기에 코드를 작성하세요.
h00 = keras.layers.Dense(units=6, activation='relu')(inputs)

h00_0 = keras.layers.Dense(units=12)(h00)
h00_1 = keras.layers.Dense(units=8, activation='relu')(h00)



h_00_1_0 = keras.layers.Dense(units=16)(h00_1)


model = keras.Model(
    inputs=inputs,
    outputs=[h00_0, h_00_1_0]
    )
```

```python
keras.utils.plot_model(model, show_shapes=True, show_layer_activations=True)
```
